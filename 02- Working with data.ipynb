```python
from __future__ import absolute_import, division, print_function, unicode_literals

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
from six.moves import urllib

import tensorflow.compat.v2.feature_column as fc

import tensorflow as tf
```

```python
x_train= pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
x_eval= pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv')
y_train=x_train.pop('survived')
y_eval=x_eval.pop('survived')
```

```python
x_train.age.hist(bins=20)
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

```python
x_train.sex.value_counts().plot(kind='barh', color=['skyblue','pink'])
plt.xlabel("Number of Passengers")
plt.ylabel("Sex")
plt.title("Distribution of Passengers by Sex")
plt.show()
```

```python
x_train['class'].value_counts().plot(kind='barh',color=['red','green','blue'])
```

```python
pd.concat([x_train,y_train],axis=1).groupby('sex').survived.mean().plot(kind='barh', color=['gray','black']).set_xlabel('% survive')
```
